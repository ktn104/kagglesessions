```python
mean = df['Total Volume'].mean() # Get the mean of the column
subtract_mean = lambda x: x - mean # Create a lambda function called subtract_mean which subtracts the mean from x
# Use the apply function from pandas to apply the subtract_mean function to every entry in Total Volume
df['Total Volume'] = df['Total Volume'].apply(subtract_mean) 
df.head()
```
